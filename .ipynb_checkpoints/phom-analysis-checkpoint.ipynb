```python
%pylab inline 
import numpy as np
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx 
sys.path.append('/Users/lordgrilo/Dropbox/')
import Holes as ho
```

```python
datadir = '../data/phom-networks/graphs/'
fs = os.listdir(datadir)
```

```python
%%sh 
ls ../data/phom-networks/results/gen
```

```python
import time
import cPickle as pk
g = {}
wgen = {}
gen = {}
for f in fs:
    try:
        print f
        now = time.time()
        g[f] = nx.read_weighted_edgelist(datadir+f)
       # weighted generators 
        gen[f] = pk.load(open('../data/phom-networks/results/gen/generators_'+f[:-6]+'_.pck'))
        wgen[f] = ho.weight_label_generators(gen[f],g[f])
    except:
        print 'missing', f
```

```python
ts = [100,200,300,400,499]
nn = 'NTGa-mem-'
ss = 'STGa-mem-'
dim = 0
fig = plt.figure(figsize=(16,4))
for i,t in enumerate(ts):
    plt.subplot(1,len(ts),i+1);
    ps = map(lambda x: x.persistence_interval(), wgen[nn+str(t)+'.edges'][dim])
    aa, bb = np.histogram(ps,10);
    plt.plot(bb[:-1], aa,'-o',label=nn+str(t))
    ps = map(lambda x: x.persistence_interval(), wgen[ss+str(t)+'.edges'][dim])
    aa, bb = np.histogram(ps,10);
    plt.plot(bb[:-1], aa,'-o',label=ss+str(t))
plt.legend(loc=0)
plt.tight_layout()

dim = 1

fig = plt.figure(figsize=(16,4))
for i,t in enumerate(ts):
    plt.subplot(1,len(ts),i+1);
    ps = map(lambda x: x.persistence_interval(), wgen[nn+str(t)+'.edges'][dim])
    aa, bb = np.histogram(ps,5);
    plt.plot(bb[:-1], aa,'-o',label=nn+str(t))
    ps = map(lambda x: x.persistence_interval(), wgen[ss+str(t)+'.edges'][dim])
    aa, bb = np.histogram(ps,5);
    plt.plot(bb[:-1], aa,'-o',label=ss+str(t))
plt.legend(loc=0)
plt.tight_layout()
```

## diagrammi di persistenza

```python
ts = [100,200,300,400,499]
nn = 'NTGa-mem-'
ss = 'STGa-mem-'
dim = 0
fig = plt.figure(figsize=(16,3))
for i,t in enumerate(ts):
    plt.subplot(1,len(ts),i+1);
    x = map(lambda x: x.start, wgen[ss+str(t)+'.edges'][dim])
    y = map(lambda x: x.end, wgen[ss+str(t)+'.edges'][dim])
    plt.plot(x,y,'o',alpha=0.3,label='ss')
    xx = map(lambda x: x.start, wgen[nn+str(t)+'.edges'][dim])
    yy = map(lambda x: x.end, wgen[nn+str(t)+'.edges'][dim])
    plt.plot(xx,yy,'o',alpha=0.3,label='nn')
    plt.xlim(0, np.max([np.max(xx), np.max(x)]))
    plt.ylim(0, np.max([np.max(yy), np.max(y)]))
plt.legend(loc=0)
plt.tight_layout()

dim = 1

fig = plt.figure(figsize=(16,3))
for i,t in enumerate(ts):
    plt.subplot(1,len(ts),i+1);
    x = map(lambda x: x.start, wgen[ss+str(t)+'.edges'][dim])
    y = map(lambda x: x.end, wgen[ss+str(t)+'.edges'][dim])
    plt.plot(x,y,'o',alpha=0.3,label='ss')
    xx = map(lambda x: x.start, wgen[nn+str(t)+'.edges'][dim])
    yy = map(lambda x: x.end, wgen[nn+str(t)+'.edges'][dim])
    plt.plot(xx,yy,'o',alpha=0.3,label='nn')
    plt.xlim(0, np.max([np.max(xx),np.max(yy), np.max(x),np.max(y)]))
    plt.ylim(0, np.max([np.max(xx),np.max(yy), np.max(x),np.max(y)]))
plt.legend(loc=0)
plt.tight_layout()
```
